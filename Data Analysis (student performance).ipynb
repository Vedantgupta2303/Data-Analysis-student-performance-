This is the general overview of the dataframe

```python
sns.pairplot(data=data,hue='gender',palette='coolwarm')
```

From the above graph we can see that the number of students getting different range of marks are more or less equal in terms of there gender. example - the marks obtained in between range of 80-100 contains approximately same amount of boys and girls.And all the plots internally co-related. for example- if we take two persons in random from these plot we can say that if person 2 have greater score in maths than person one thean on an average we can predict that person 2 reading score would be greater than that of person 1.

```python
sns.jointplot(x='math score',y='reading score',data=data,kind='hex',height=5,ratio=2)
plt.show()
sns.jointplot(x='math score',y='writing score',data=data,kind='hex',height=5,ratio=2) 
plt.show()
sns.jointplot(x='writing score',y='reading score',data=data,kind='hex',height=5,ratio=2)
plt.show()
```

From the above three plots we can see that people are scoring more between 60-80 marks in almost all the subjects.

```python
sns.barplot(y='parental level of education',x='math score',data=data)
plt.show()
sns.barplot(y='parental level of education',x='writing score',data=data)
plt.show()
sns.barplot(y='parental level of education',x='reading score',data=data)
plt.show()
```

From all the above plots we can see that the students whose parental level of education is master's degree score more in every subject and so on for the rest.

```python
sns.boxplot(x="math score", y="race/ethnicity", data=data,palette='rainbow')
plt.show()
sns.boxplot(x="writing score", y="race/ethnicity", data=data,palette='rainbow')
plt.show()
sns.boxplot(x="reading score", y="race/ethnicity", data=data,palette='rainbow')
plt.show()
```

1) In the above plot we can see that the student of group E is altmost good in every subjects than that of students of other groups... 
2) Student belonging to group c have always got the lowest marks in all the scores.... 
3) In al the subject students of group E and group D has scored full marks... 
4) Group A student are among one of the poorest scorer all the time

```python
sns.barplot(x='gender',y='math score',data=data)
plt.show()
sns.barplot(x='gender',y='writing score',data=data)
plt.show()
sns.barplot(x='gender',y='reading score',data=data)
plt.show()
```

From the above plots we can see that the mean marks of male is more in that of maths and writing but the mean marks of females is greater in reading than males. But as we can see that the diferrence is very little in all the 3 cases so the gender is not directly affecting the scores to a great extent.

```python
sns.barplot(x='lunch',y='writing score',data=data,palette='coolwarm')
plt.show()
sns.barplot(x='lunch',y='reading score',data=data,palette='coolwarm')
plt.show()
sns.barplot(x='lunch',y='math score',data=data,palette='coolwarm')
plt.show()
```

All the 3 plots above shows that we can the students whose lunch is standard has a average value of score better that taht of free/reduced lunch in all the subjects 

```python
sns.barplot(x='test preparation course',y='math score',data=data,palette='rainbow')
plt.show()
sns.barplot(x='test preparation course',y='reading score',data=data,palette='rainbow')
plt.show()
sns.barplot(x='test preparation course',y='writing score',data=data,palette='rainbow')
plt.show()
```

The student who has completed the test preparation course have scored good in all the subject... This shows that student requires test preparation course to perform better in exam 

```python
ax = sns.violinplot(x="parental level of education", y="math score", data=data,hue='gender',split=True,palette='Set1')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
ax = sns.violinplot(x="parental level of education", y="reading score", data=data,hue='gender',split=True,palette='Set1')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
ax = sns.violinplot(x="parental level of education", y="writing score", data=data,hue='gender',split=True,palette='Set1')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
plt.tight_layout()
```

In all the cases metioned above we can see that in a girls are good at reading and writing what so ever be there parental level of education than boy but in case of math the vice-versa is true.
We can also come with a fact that the boys whose parental level of education is master's degree has a better mean value than that of girls in maths , and the girls who have there parental education of bachelor's degree has a good average in maths.
In all other subjects we can also interpret the same thing about boys,girls and there parental level of education as metioned above.

```python
sns.swarmplot(x="gender", y="math score", data=data,hue='test preparation course')
plt.show()
sns.swarmplot(x="gender", y="reading score", data=data,hue='test preparation course')
plt.show()
sns.swarmplot(x="gender", y="writing score", data=data,hue='test preparation course')
plt.show()
```

From the above plot we can see that the number of student completed there test preperation are less than that of those who didn't completed the course and we can also notice the fact the girls are good at writing and reading than boys as they have scored more than boys in the respective subjects but boys are good at math than that of girls.

```python
ax=sns.swarmplot(x="parental level of education", y="math score", data=data,hue='race/ethnicity',palette='rainbow',dodge=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
ax=sns.swarmplot(x="parental level of education", y="reading score", data=data,hue='race/ethnicity',palette='rainbow',dodge=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
ax=sns.swarmplot(x="parental level of education", y="writing score", data=data,hue='race/ethnicity',palette='rainbow',dodge=True)
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

From the above plot we can see that the people belonging to Group C are there in a huge number though the student of group E are scoring the best.
